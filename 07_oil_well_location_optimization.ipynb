## Обучение и проверка модели

#### Объединять ли датасеты регионов месторождений в один для обучения модели или обучать модель на каждом регионе индивидуально?

- В нашем случае, предполагаю, что каждый регион представляет собой разные геологические особенности, а также разные характеристики месторождений, и считаю, что наилучшим подходом будет обучение модели на каждом регионе индивидуально.

- Обучение модели на каждом регионе отдельно позволит модели лучше улавливать уникальные особенности каждого региона и адаптироваться к ним. Это может привести к более точным прогнозам для каждого региона.

#### Напишем функцию model_training_and_validation:
- Которая будет иметь на вход аргумент data - с данными.
- Разбивать данные по признакам.
- Разбивать данные признаков на обучающую и валидационную выборки.
- Проводить масштабирование признаков. 
   - Хоть в задание ничего про это не сказано, и мы не знаем какие конкретно признаки имеют месторождения, однако их масштаб может отличаться. Например это могут быть признаки глубины месторождения и возраст его возникновения, маловероятно, что признаки определяющие запасы месторождений имеют одинаковый диапазон масштабов, ведь, вероятнее всего, обозначают соверешенно разные вещи. 
   - P.S: я увидел, при дальнейшим написание проекта, что это особо не влияет, но решил оставить.
- Создавать экземпляр модели, обучать его на обучающий выборке.
- Получать предсказания на на валидационной выборке.
- Вычислять среднее значение предсказанных запасов.
- Вычислять RMSE модели.
- Возвращать обученную на тренировочной выборке модель, значения предсказаний модели и правильных ответов на валидационной выборке, средний запас предсказанного сырья и RMSE модели.

```python
def model_training_and_validation(data):
    #Разобьем данные по признакам
    features = data.drop(['product'], axis=1)
    target = data['product']
    #Разобъем данные признаков на обучающую и валидационную выборки
    features_train, features_valid, target_train, target_valid = train_test_split(
                                                                            features,
                                                                            target,
                                                                            test_size=0.75,
                                                                            random_state=12345
                                                                            )
    
    #Проведем масштабирование признаков с помощью StandardScaler
    #cоздадим экземпляр scaler
    scaler = StandardScaler()
    #обучим на тренировочной выборке
    scaler.fit(features_train)
    #масштабируем тренировочную выборку
    features_train = scaler.transform(features_train)
    #масштабируем тестовую выборку
    features_valid = scaler.transform(features_valid)
    
    #Создадим экземпляр модели
    model = LinearRegression()
    #Обучим модель на обучающей выборке
    model.fit(features_train, target_train)
    #Получим и сохраним в переменную предсказания модели на валидационной выборке
    predictions_valid = model.predict(features_valid)
    
    #Вычислим среднее значение предсказанных запасов
    mean_predictions_valid_volume = np.mean(predictions_valid)
    
    #Получим среднеквадратическую ошибку (MSE)
    mse = mean_squared_error(target_valid, predictions_valid)
    #Вычислить RMSE, вычислив квадратный корень из MSE
    rmse = np.sqrt(mse)
    
    #Вернем обученную на тренировочной выборке модель
    #Вернем значения предсказаний модели и правильных ответов на валидационной выборке 
    #Вернем средний запас предсказанного сырья и RMSE
    return model, predictions_valid, target_valid, mean_predictions_valid_volume, rmse
```

#### Применим функцию поочередно к датасетам регионов и сохраним возвращенные ею значения в соответствующии переменные.

```python
#Применим функцию к каждому датасету и запишем выводимые ее значения в переменную для каждого датасета
result_model_geo_1 = model_training_and_validation(df_geo_1)
result_model_geo_2 = model_training_and_validation(df_geo_2)
result_model_geo_3 = model_training_and_validation(df_geo_3)
```

#### Из переменных с результатами работы функции по регионам вытащим каждое значения и запишим в соответствующую переменную

```python
def calculate_profit(target, predictions, development_points, barrel_price, budget, income_per_unit):
    #Преобразуем массива NumPy в Series Pandas
    predictions_series = pd.Series(predictions)

    #Получим предсказания для выбранных индексов
    exploration_points_predictions = predictions.sort_values().tail(development_points)
    #Рассчитаем прибыль для 200 наилучших месторождений из выбранных
    profit = (target.iloc[exploration_points_predictions.index].sum() * income_per_unit) - budget
    
    return profit
```

#### Напишим функцию для подсчета рисков и прибыли.
- Функция будет применять технику Bootstrap с 1000 выборками. 
- Считать прибыль с помощью функции calculate_profit, написанной выше.
- Находить среднюю прибыль.
- Находить 95%-й доверительный интервал.
- Считать риск убытков.

```python
def calculate_with_bootstrap_risk_and_profit(target, predictions, exploration_points, development_points, barrel_price, budget, income_per_unit, bootstrap_samples=1000):
    #Создадим список с профитами
    profits = []
    #Установим seed для генератора случайных чисел для воспроизводимости результатов
    state = np.random.RandomState(12345)
    #Преобразуем массива NumPy в Series Pandas
    predictions = pd.Series(predictions)
    
    #Напишем цикл для Bootstrap с 1000 выборками
    for sample in range(bootstrap_samples):
        #Выберем случайные месторождения для разведки региона
        exploration_points_random_sample = predictions.sample(n=exploration_points, 
                                                              random_state=state,
                                                              replace = True)
        
        #Рассчитаем прибыль для 200 наилучших месторождений из выбранных
        profit = calculate_profit(target, 
                                  exploration_points_random_sample, 
                                  development_points, 
                                  barrel_price, budget, 
                                  income_per_unit)
        
        #Добавим значение в список с профитами
        profits.append(profit)
    
    #Запишем список профитов как Pandas Series для удобства
    profits = pd.Series(profits)
    profits = profits.astype(float)
    
    #Найдем среднюю прибыль
    mean_profit = profits.mean()
    
    #Найдем 95%-й доверительный интервал
    lower_bound = np.percentile(profits, 2.5)
    upper_bound = np.percentile(profits, 97.5)
    
    #Найдем риск убытков
    loss_probability = (profits < 0).mean() * 100

    #Вернем список с профитами, среднюю прибыль, 95%-й доверительный интервал и риск убытков
    return profits, mean_profit, lower_bound, upper_bound, loss_probability
```
