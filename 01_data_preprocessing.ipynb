**Task 5. There are missing values in two columns. One of them is `days_employed`. The missing values in this column will be handled in the next step. The other column with missing values is `total_income`, which contains information about income. Since income is most strongly influenced by the type of employment, the missing values in this column should be filled with the median value for each employment type from the `income_type` column. For example, if a person with the employment type `employee` has a missing value in the `total_income` column, it should be replaced with the median income calculated among all records with the same employment type.**

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
```

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.

**Например, кредитополучателю с доходом 25000 нужно назначить категорию `'E'`, а клиенту, получающему 235000, — `'B'`. Используйте собственную функцию с именем `categorize_income()` и метод `apply()`.**

---

**Task 16. Based on the ranges specified below, create a column `total_income_category` in the `data` DataFrame with the following categories:**

- 0–30,000 — `'E'`;
- 30,001–50,000 — `'D'`;
- 50,001–200,000 — `'C'`;
- 200,001–1,000,000 — `'B'`;
- 1,000,001 and above — `'A'`.

**For example, a borrower with an income of 25,000 should be assigned category `'E'`, while a client with an income of 235,000 should be assigned category `'B'`. Use a custom function named `categorize_income()` together with the `apply()` method.**

```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

**Task 18. Create a function that, based on the values in the `purpose` column, generates a new column `purpose_category` with the following categories:**

- `'car-related operations'`,
- `'real estate operations'`,
- `'wedding expenses'`,
- `'education expenses'`.

**For example, if the `purpose` column contains the substring `'for buying a car'`, the corresponding value in the `purpose_category` column should be `'car-related operations'`.**

**Use a custom function named `categorize_purpose()` together with the `apply()` method. Examine the data in the `purpose` column and determine which substrings can help correctly identify the appropriate category.**

```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

#### 3.1 Есть ли зависимость между количеством детей и возвратом кредита в срок?

#### 3.1 Is there a relationship between the number of children and the timely repayment of a loan?

```python
pivot_table = pd.pivot_table(data, 
                             index=['children'], values=['debt'], aggfunc=['count', 'sum', 'mean'],  fill_value=0)
print(pivot_table)
```

**Вывод:** Для выборки возьмем самые многочисленные категории количества детей: 0, 1, 2; это необходимо сделать для учета несбалинсированности выборок — чем меньше количество данных, тем больше влияние на них случайности и тем ниже надежность выводов. Например, если подбросить монетку 5 раз, может выпасть 5 орлов подряд, но если подбросить монетку 1000 раз, количество орлов и решек будет стремиться к одинаковому соотношению.  
По выбранным категориям можно сказать, что существует зависимость между уменьшением вероятности своевременного возврата кредита и увеличением количества детей. Этот вывод выглядит логичным: чем больше детей, тем выше финансовая нагрузка заемщика, при этом доход, как правило, не увеличивается.

---

**Conclusion:** For the analysis, we consider the most common categories for the number of children: 0, 1, and 2. This is done to account for sample imbalance — the smaller the number of observations in a category, the greater the influence of randomness and the lower the reliability of conclusions. For example, if a coin is flipped 5 times, it may land on heads all 5 times; however, if it is flipped 1,000 times, the number of heads and tails will tend toward an equal proportion.  

Based on the selected categories, we can conclude that there is a relationship between the probability of timely loan repayment and the number of children: as the number of children increases, the likelihood of repaying the loan on time decreases. This result is intuitive: the more children a borrower has, the greater the financial burden, while income typically does not increase accordingly.

#### 3.6 Explain why filling missing values with the median is the best solution for quantitative variables.

Я слегка улучшил стиль, чтобы текст звучал более академично и естественно на английском, но при этом сохранил вашу мысль.

*Ответ:* Медианное значение лучшее решение, потому что оно более реалистично отражает распределение данных, в отличие от среднего значения, которое может быть искажено экстремальными значениями.

Медиана представляет собой значение, относительно которого половина наблюдений находится выше, а половина — ниже. Это означает, что при достаточно большом количестве наблюдений аномальные значения (включая реальные, но сильно отличающиеся от большинства) оказывают значительно меньшее влияние на итоговый показатель. В результате медиана лучше отражает типичное значение в выборке и позволяет получить более устойчивую оценку при заполнении пропусков.

---

*Answer:* Using the median is often a better solution for filling missing values in quantitative variables because it provides a more realistic representation of the data distribution compared to the mean, which can be strongly affected by extreme values.

The median represents the value that divides the dataset into two equal halves: half of the observations are above it and half are below it. This means that when the dataset is sufficiently large, outliers — including valid but unusually large or small values — have much less influence on the result. As a result, the median better reflects the typical value in the dataset and provides a more robust estimate when imputing missing values.
